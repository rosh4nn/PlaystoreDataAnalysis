```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('CleanedPlaystoreData.csv')
```

```python
df.head()
```

```python
cat_wise = df.groupby('Category')['Category'].count().reset_index(name='cat')
cat_wise = cat_wise.sort_values('cat')

plt.barh(cat_wise['Category'],cat_wise['cat'],height=0.79,edgecolor='k',color='steelblue')

plt.title('App-Category Report')
plt.ylabel('Categories')
plt.xlabel('No.of Apps')

plt.tight_layout()
plt.show()
```

```python
df['Price_Category'].unique()
```

```python
pc = df.groupby('Price Category')['Price Category'].count().reset_index(name='price_cat')

bars = plt.bar(pc['Price Category'],pc['price_cat'],color='steelblue',edgecolor='k')
for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2,
             bar.get_height(),
             int(bar.get_height()),
             ha='center',
             va='bottom'
             )

plt.title('Category-Wise Report')
plt.xlabel('Price Category')
plt.ylabel('No.of Applications')

plt.tight_layout()
plt.show()
```

```python
df_type = df.groupby('Type')['Type'].count().reset_index(name='Count')

plt.pie(df_type['Count'],labels=df_type['Type'],autopct='%1.2f%%',explode=[0,0.1],colors=['steelblue','purple'],startangle=90)
plt.title('Types of Applications')
plt.show()
```

```python
df['Last Updated Year'].unique()
```

```python
df_type = df.groupby('Content Rating')['Category'].count().reset_index(name='Count')

bars = plt.bar(df_type['Content Rating'],df_type['Count'],color='steelblue',edgecolor='k')

for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2,
            bar.get_height(),
            int(bar.get_height()),
            ha='center',
            va='bottom')
plt.tight_layout()
plt.show()
```

```python
last_updated_app_year = df.groupby('Last Updated Year')['App'].count().reset_index(name='Count')

bars = plt.bar(last_updated_app_year['Last Updated Year'],last_updated_app_year['Count'],color='steelblue',edgecolor='k')

plt.xticks(last_updated_app_year['Last Updated Year'],[2010,2011,2012,2013,2014,2015,2016,2017,2018])
for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height(),
        int(bar.get_height()),
        ha='center',
        va='bottom'
    )
plt.show()
```

```python
plt.hist(df['Rating'],bins=5,rwidth=0.60,color='gold',edgecolor='k')

plt.title('Rating Report')
plt.xlabel('Rating')
plt.ylabel('No.of Applications')

plt.tight_layout()
plt.show()
```

```python
app_installed = df.groupby('Category')['Installs'].sum().reset_index(name='Sum')
app_installed = app_installed.sort_values(by='Sum')

plt.barh(app_installed['Category'],app_installed['Sum'],height=0.79,color='steelblue',edgecolor='k')
plt.yticks(app_installed['Category'])

plt.tight_layout()
plt.show()
```
